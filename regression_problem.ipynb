```python
def find_no_of_outliers(list_of_features):
    for feature in list_of_features:
        Q1 = df[feature].quantile(q=0.25)
        Q3 = df[feature].quantile(q=0.75)
        IQR = Q3 - Q1
        upper_limit = Q3 + IQR*1.5
        lower_limit = Q1 - IQR*1.5
        colm = df[(df[feature]<lower_limit) | (df[feature]>upper_limit)]
        no_of_outliers = colm.shape[0]
        print('no of outiler %s : %d'%(feature,no_of_outliers))

    
    
```

since our data is not that big we cant drop the outliers so lets see if we can change ta outlers with its column median value

```python
outlier_column = ['CRIM', 'ZN', 'CHAS', 'RM','DIS',
       'PTRATIO', 'B', 'LSTAT', 'price']
```

```python
def return_index_of_outler(column):
    Q1 = df[column].quantile(q=0.25)
    Q3 = df[column].quantile(q=0.75)
    IQR = Q3 - Q1
    upper_limit = Q3 + IQR*1.5
    lower_limit = Q1 - IQR*1.5
    ls = df.index[(df[column]<lower_limit) | (df[column]>upper_limit)]
    return ls
```

```python
def change_outlier_mean(outlier_column):
    for feature in outlier_column:
        ls = return_index_of_outler(feature)
        for i in ls:
            df[feature][i] = df[feature].median()
```

```python
change_outlier_mean(outlier_column)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns='price',axis=1)
y = df['price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict = model.predict(x_test)
```

```python
score = model.score(x_test,y_test)
score
```

the score is not so good so good so lets use 

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,predict)
```

```python
mean_squared_error(y_test,predict)
```

```python
plt.scatter(y_test,predict)
plt.xlabel("actual price")
plt.ylabel("predict")
plt.plot(y_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2 = RandomForestRegressor()
```

```python
predict1 = model2.predict(x_test)
```
